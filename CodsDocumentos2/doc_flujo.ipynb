```python
import pandas as pd
import numpy as np
import tabula
```

```python
pdf = "../Inputs/Docs2LaFavorita/flujo.pdf"
```

```python
firstPage = tabula.read_pdf(pdf, pages=1, area=(99,20, 816, 573), lattice=True, output_format="dataframe",encoding='latin-1')
second = tabula.read_pdf(pdf, pages=2, area=(20,20,525,575), lattice=True, output_format="dataframe",encoding='latin-1')
Header = tabula.read_pdf(pdf, pages=1, area=(19,20,101,574), lattice=True, output_format="dataframe",encoding='latin-1')
```

```python
daf1 = firstPage[0]
daf2 = second[0]
dafHeader = Header[0]
```

```python
daf1.head(8)
```

```python
daf1.rename(inplace=True, columns={"ESTADO DE FLUJO EFECTIVO POR EL MÉTODO DIRECTO":"CUENTA",
                                "Unnamed: 1":"CODS", "Unnamed: 2":"SALDOS BALANCE EN USD"})

daf1.drop(["CODS"], axis=1, inplace=True)
daf1.head(8)
```

```python
daf1.drop([0],inplace=True)
daf1.head(8)
```

```python
daf2.head(7)
```

```python
daf2.rename(inplace=True, columns={"ESTADO DE FLUJO EFECTIVO POR EL MÉTODO DIRECTO":"CUENTA",
                                "Unnamed: 1":"CODS", "Unnamed: 2":"SALDOS BALANCE EN USD"})

daf2.drop(["CODS"], axis=1, inplace=True)
daf2.drop([0],inplace=True)
daf2.head(8)
```

```python
print(daf1.info())
print(daf2.info())
```

```python
dafData = pd.concat([daf1, daf2]) 
```

```python
dafData
```

```python
texto = dafData["CUENTA"].to_list()
valores_list = dafData["SALDOS BALANCE EN USD"].to_list()
```

```python
valores = np.array(valores_list).reshape(1,-1)
df2=pd.DataFrame(valores, columns=texto)
```
