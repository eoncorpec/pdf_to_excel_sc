```python
import pandas as pd
import numpy as np
import tabula
```

```python
pdf = "../Inputs/Docs2LaFavorita/resultado_integral.pdf"
```

```python
header = tabula.read_pdf(pdf, pages=1, area=(20,19,100,575), lattice=True, output_format="dataframe", encoding="latin-1")
first = tabula.read_pdf(pdf, pages=1, area=(99, 20, 815, 572), lattice=True, output_format="dataframe", encoding="latin-1")
second = tabula.read_pdf(pdf, pages=2, area=(20,20, 821,575), lattice=True, output_format="dataframe", encoding="latin-1")
third = tabula.read_pdf(pdf, pages=3, area=(20,20,821,575), lattice=True, output_format="dataframe", encoding="latin-1")
fourth = tabula.read_pdf(pdf, pages=4, area=(20,20,821,575), lattice=True, output_format="dataframe", encoding="latin-1")
fifth = tabula.read_pdf(pdf, pages=5, area=(20,19,499,575), lattice=True, output_format="dataframe", encoding="latin-1")

```

```python
dafHeader = header[0]
daf1 = first[0]
daf2 = second[0]
daf3 = third[0]
daf4 = fourth[0]
daf5 = fifth[0]
```

```python
daf4.rename(inplace=True, columns={"ESTADO DE RESULTADO INTEGRAL":"CUENTA",
                                "Unnamed: 1":"CODS", "Unnamed: 2":"VALOR EN USD"})

daf4.drop(["CODS"], axis=1, inplace=True)
daf4.drop([0], inplace=True)
daf4.head(8)
```

```python
dafData = pd.concat([daf1,daf2, daf3, daf4, daf5])
```

```python
texto = dafData["CUENTA"].to_list()
valores_list = dafData["VALOR EN USD"].to_list()
```

```python
valores = np.array(valores_list).reshape(1,-1)
datos = pd.DataFrame(valores, columns=texto)
```
