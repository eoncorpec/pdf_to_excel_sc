```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("./Outputs/Doc1/Improvecontinued.xlsx")
```

```python
df.head(20)
```

```python
df.dropna(axis=0, inplace=True)
```

```python
columnas = []
for index, row in df.iloc[:].iterrows():
    columnas.append(row.tolist())

print(columnas[0:20])
```

```python
columnas_aplanadas = []

for elemento in columnas:
    if isinstance(elemento, list):
        columnas_aplanadas.extend(elemento)
    else:
        columnas_aplanadas.append(elemento)

columnas_aplanadas = [str(item) for item in columnas_aplanadas]
```

```python
import re

def getCuentaTexto(elemento):
    match = re.match(r'^([^\d]+)', elemento)
    return match.group(1) if match else elemento

columnas_texto = [getCuentaTexto(item) for item in columnas_aplanadas]
```

```python
print(columnas_texto)
```

```python
columnas_aplanadas
```

```python
def posicionInicioCodigo(elemento) -> list:
    positions = [re.search(r'\d', k).start() if re.search(r'\d', k) else -1 for k in elemento[1:]]
    return positions

positions = posicionInicioCodigo(columnas_aplanadas)
```

```python
columnas_codigos = []
for k,pos in zip( range(1,len(columnas_aplanadas)), range(0,len(positions)) ):
    columnas_codigos.append( columnas_aplanadas[k][positions[pos] : ] )

columnas_codigos[0:15]
print(len(columnas_codigos))
```

```python
codigos = []
for k in range(0, len(columnas_codigos)):
    codigos.append(columnas_codigos[k].split(" ")[0])

codigos[0:35]
```

```python
columnas_codigos[170:]
```

```python
columnas_codigos_y_valores = [
    item for item in columnas_codigos if re.search(r'^\d', item)
]
```

```python
print(columnas_codigos_y_valores)
print(len(columnas_codigos_y_valores))
```

```python
valoresUSD = []
for k in range(0, len(columnas_codigos_y_valores)):
    if( len(columnas_codigos_y_valores[k]) > 1 ):
        valoresUSD.append(columnas_codigos_y_valores[k].split(" ")[-1])
    elif columnas_codigos_y_valores[k] == "0 ACCIONISTAS PARA FUTURA CAPITALIZACION 302 0.00":
        valoresUSD.append(columnas_codigos_y_valores[k].split(" ")[-1])
    else:
        valoresUSD.append(columnas_codigos_y_valores[k])

valoresUSD
```
